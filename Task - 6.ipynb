```python
from tkinter import *
import random
from tkinter import messagebox

root = Tk()
root.title('Billing Slip')
root.geometry('1280x720')
bg_color = '#a6e7ff'

# Variables
c_name = StringVar()
c_phone = StringVar()
item = StringVar()
Price = IntVar()
Qty = IntVar()
bill_no = StringVar()
x = random.randint(1000, 9999)
bill_no.set(str(x))

global l
l = []

# Function
def welcome():
    textarea.delete(1.0, END)
    textarea.insert(END, "\t\t Welcome Nayana Retails")
    textarea.insert(END, f'\n\nBill Number: \t\t{bill_no.get()}')
    textarea.insert(END, f'\n Customer Name:\t\t{c_name.get()}')
    textarea.insert(END, f'\n Phone Number:\t\t{c_phone.get()}')
    textarea.insert(END, f"\n================================================")
    textarea.insert(END, f'\n Product\t\t             QTY\t\t               Price')
    textarea.insert(END, f"\n================================================\n")
    textarea.configure(font='arial 14 bold')

def additm():
    n = Price.get()
    m = Qty.get() * n
    l.append(m)
    if item.get() == '':
        messagebox.showerror('Error', 'Please Enter the item')
    else:
        textarea.insert(END, f'\n{item.get()}\t\t{Qty.get()}\t\t{m}')

def gbill():
    if c_name.get() == '' or c_phone.get() == '':
        messagebox.showerror('Error', 'Customer details not mentioned')
    else:
        welcome()  
        for i in range(len(l)):
            textarea.insert(END, f'\n{item.get()}\t\t{Qty.get()}\t\t{l[i]}')
        
        textarea.insert(END, f"\n================================================")
        textarea.insert(END, f"\nTotal Payable Amount:\t\t\t{sum(l)} ")
        textarea.insert(END, f"\n================================================")
    savebill()

def savebill():
    op = messagebox.askyesno('Save bill', 'Do you want to save the bill?')
    if op > 0:
        bill_details = textarea.get(1.0, END)
        file_path = "C:/Users/JIGISHA NAYANA/OneDrive/Desktop/bills/" + str(bill_no.get()) + ".txt"
        with open(file_path, 'w') as f1:
            f1.write(bill_details)
        messagebox.showinfo('Saved', f'Bill no: {bill_no.get()} saved successfully')
    else:
        return


def clear():
    c_name.set('')
    c_phone.set('')
    item.set('')
    Price.set(0)
    Qty.set(0)
    l.clear()
    welcome()

def exit():
    op = messagebox.askyesno('Exit', 'Do you really want to exit')
    if op > 0:
        root.destroy()

# Top section
title = Label(root, text='Billing Software', bg=bg_color, fg='#0b016b', font=('times new romman', 25, 'bold'), relief=GROOVE, bd=12, padx=10, pady=5)
title.pack(fill=X)

# Customer details
F1 = LabelFrame(root, text='[ Customer Details ]', bg=bg_color, fg='black', font=('times new romman', 18, 'bold'), relief=GROOVE, bd=10, padx=10, pady=5)
F1.place(x=0, y=65, relwidth=1)

cname_lbl = Label(F1, text='Customer Name : ', bg=bg_color, fg='#0b016b', font=('times new romman', 14, 'bold'), relief=GROOVE, bd=0)
cname_lbl.grid(row=0, column=0, padx=10, pady=5)

cname_txt = Entry(F1, width=15, font='arial 15 bold', relief=SUNKEN, textvariable=c_name)
cname_txt.grid(row=0, column=1, padx=10, pady=5)

cphone_lbl = Label(F1, text='Phone No :            ', bg=bg_color, fg='#0b016b', font=('times new romman', 14, 'bold'), relief=GROOVE, bd=0)
cphone_lbl.grid(row=1, column=0, padx=10, pady=5)

cphone_txt = Entry(F1, width=15, font='arial 15 bold', relief=SUNKEN, textvariable=c_phone)
cphone_txt.grid(row=1, column=1, padx=10, pady=5)

# Product details
F2 = LabelFrame(root, text='[ Product Details ]', bg=bg_color, fg='black', font=('times new romman', 18, 'bold'), relief=GROOVE, bd=10, padx=10, pady=5)
F2.place(x=0, y=190, width=630, height=450)

itm = Label(F2, text='Product Name : ', bg=bg_color, fg='#0b016b', font=('times new romman', 14, 'bold'), relief=GROOVE, bd=0)
itm.grid(row=0, column=0, padx=50, pady=5)
itm_txt = Entry(F2, width=20, font='arial 15 bold', textvariable=item)
itm_txt.grid(row=0, column=1, padx=50, pady=5)

price = Label(F2, text='Product Price :  ', bg=bg_color, fg='#0b016b', font=('times new romman', 14, 'bold'), relief=GROOVE, bd=0)
price.grid(row=1, column=0, padx=50, pady=5)
price_txt = Entry(F2, width=20, font='arial 15 bold', textvariable=Price)
price_txt.grid(row=1, column=1, padx=50, pady=5)

qty = Label(F2, text='Quantity :            ', bg=bg_color, fg='#0b016b', font=('times new romman', 14, 'bold'), relief=GROOVE, bd=0)
qty.grid(row=2, column=0, padx=50, pady=5)
qty_txt = Entry(F2, width=20, font='arial 15 bold', textvariable=Qty)
qty_txt.grid(row=2, column=1, padx=50, pady=5)

# Buttons
btn1 = Button(F2, text='Add item', font='arial 15 bold', padx=5, pady=5, bg='#00b7eb', bd=0, width=12, command=additm)
btn1.grid(row=3, column=0, padx=50, pady=30)

btn2 = Button(F2, text='Generate Bill', font='arial 15 bold', padx=5, pady=5, bg='#00b7eb', bd=0, width=12, command=gbill)
btn2.grid(row=3, column=1, padx=50, pady=30)

btn3 = Button(F2, text='Clear', font='arial 15 bold', padx=5, pady=5, bg='#00b7eb', bd=0, width=12, command=clear)
btn3.grid(row=4, column=0, padx=50, pady=20)

btn4 = Button(F2, text='Exit', font='arial 15 bold', padx=5, pady=5, bg='#00b7eb', bd=0, width=12, command=exit)
btn4.grid(row=4, column=1, padx=50, pady=20)

# Bill area
F3 = Frame(root, relief=GROOVE, bd=10)
F3.place(x=670, y=190, width=570, height=450)

bill_title = Label(F3, text='Invoice Preview', font='arial 15 bold', relief=GROOVE, bd=7).pack(fill=X)
scroll_y = Scrollbar(F3, orient=VERTICAL)
textarea = Text(F3, yscrollcommand=scroll_y.set)
scroll_y.pack(side=RIGHT, fill=Y)
scroll_y.config(command=textarea.yview)
textarea.pack()
welcome()

root.mainloop()
```
